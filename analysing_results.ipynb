```python
import pandas as pd
```

```python
pseudo = pd.read_csv("./Pseudo_stochastic_results/random_5.csv")
pseudo_weighted = pd.read_csv("./Pseudo_stochastic_results/weighted_5.csv")
robust_0 = pd.read_csv("./robust_results/robust_0.csv", header = None)
robust_1 = pd.read_csv("./robust_results/robust_1.csv", header = None)
robust_2 = pd.read_csv("./robust_results/robust_2.csv", header = None)
robust_3 = pd.read_csv("./robust_results/robust_3.csv", header = None)
sequential = pd.read_csv("./Sequential_results/sequential.csv")
robust_0.columns = pseudo.columns
robust_1.columns = pseudo.columns
robust_2.columns = pseudo.columns
robust_3.columns = pseudo.columns
sequential.columns = pseudo.columns
```

```python
blue_utilty = pd.concat([sequential["Blue_Utility"], robust_0["Blue_Utility"],  pseudo_weighted["Blue_Utility"],pseudo["Blue_Utility"], robust_1["Blue_Utility"], robust_2["Blue_Utility"],  robust_3["Blue_Utility"]], axis = 1)
red_utilty = pd.concat([sequential["Red_Utility"], robust_0["Red_Utility"], pseudo_weighted["Red_Utility"],pseudo["Red_Utility"], robust_1["Red_Utility"],robust_2["Red_Utility"],robust_3["Red_Utility"]], axis = 1)
```

```python
blue_utilty
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10, 10))
blue_utilty.boxplot(positions = [0, 2, 4, 6, 8, 10, 12], color = dict(boxes="blue", whiskers="blue", medians="cyan", caps="blue"), ax = ax, patch_artist = True, return_type = 'both')
red_utilty.boxplot(positions = [1, 3, 5, 7, 9, 11, 13], color = dict(boxes="pink", whiskers="pink", medians="red", caps="red"), ax = ax, patch_artist = True, return_type = 'both')


ax.set_xticklabels([
   plt.text(0, 0, "Sequential"),
 plt.text(0, 0, 'Simultaneous'),
 plt.text(2, 0, 'Pseudo Stochastic \n Weighted 5'),
plt.text(4, 0, 'Pseudo Stochastic \n Top 5'),
  plt.text(6, 0, r'Robust $\Gamma = 1$'),
   plt.text(8, 0, r'Robust $\Gamma = 2$'),
   plt.text(10, 0, r'Robust $\Gamma = 3$'),
 plt.text(1, 0, ''),
 plt.text(3, 0, ''),
 plt.text(5, 0, ''),
  plt.text(1, 0, ''),
 plt.text(3, 0, ''),
 plt.text(5, 0, ''),
  plt.text(7, 0, '')
 ], rotation = 45)

ax.axvline(1.5, color = 'black', linestyle = '--')
ax.axvline(3.5, color = 'black', linestyle = '--')
ax.axvline(5.5, color = 'black', linestyle = '--')
ax.axvline(7.5, color = 'black', linestyle = '--')
ax.axvline(9.5, color = 'black', linestyle = '--')
ax.axvline(11.5, color = 'black', linestyle = '--')

ax.set_title("Utility of blue (friendly) and red (enemy) players under different optimization methods", fontsize = 12)

fig.savefig("results.png")
```
